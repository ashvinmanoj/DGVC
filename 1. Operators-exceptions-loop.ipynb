### **RANGE() function** 
**The range() function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and stops before a specified number.**



```python
print(range(5))

print("INCREMENTED FOR LOOP WITH STEP=1")
for i in range(0,5,1):
    print(i)

print("INCREMENTED FOR LOOP WITH STEP=-1")
for i in range(0,5,-1):
    print(i)
    
print("DECREMENTED FOR LOOP")
for i in range(4, -1, -1):
    print(i)
```

### FOR loop with break

```python
lis = [1, 2, 3, 4, 5]
i = 0
 
for j in range(len(lis)): 
    # Terminating condition for i
    if(i >= len(lis)):
        break

    print("Value of j is : ", j)
    print("Value of i is : ", i)
    i += 2
```

### Difference between iterable and iterator

Iterable is an object, that one can iterate over. It generates an Iterator when passed to **iter()** method. An iterator is an object, which is used to iterate over an iterable object using the __next__() method. Iterators have the __next__() method, which returns the next item of the object.

**Note**: Every iterator is also an iterable, but not every iterable is an iterator in Python.

```python
s="GFG"
s=iter(s)
print(s)
print(next(s))
print(next(s))
print(next(s))
```

### Try the iterator using a list object instead of a string

### Try Except to identify errors

The **try** block lets you test a block of code for errors.

The **except** block lets you handle the error.

The **else** block lets you execute code when there is no error.

The **finally** block lets you execute code, regardless of the result of the try- and except blocks.

```python
print(x)
```

```python
try:
    print(x)
except:
    print("An exception occurred")
```

**Print one message if the try block raises a NameError and another for other errors**

```python
try:
    print(x)
except NameError:
    print("Variable x is not defined")
except:
    print("Something else went wrong")
```

**Else**

*You can use the else keyword to define a block of code to be executed if no errors were raised:*

In the next example, the try block does not generate any error

```python
try:
    print(x)
except:
    print("Something went wrong")
else:
    print("Nothing went wrong")
```

```python
try:
    print(10)
except:
    print("Something went wrong")
else:
    print("Nothing went wrong")
```

**Finally**

*The finally block, if specified, will be executed regardless if the try block raises an error or not.*

```python
try:
    print(x)
except:
    print("Something went wrong")
finally:
    print("The 'try except' is finished")
```

### Scope of variables and functions
1. **Local Scope**: A variable created inside a function is available inside that function

2. **Global Scope**: A variable created in the main body of the Python code is a global variable and belongs to the global scope

#### Local Scope examples

Try to print the value of x outside the function

```python
def myfunc():
    x = 300
    print(x)

myfunc()
```

#### Function Inside Function

The variable x is not available outside the function, but it is available for any function inside the function:

```python
def myfunc():
    x = 300
    def myinnerfunc():
        print(x)
    
    myinnerfunc()

myfunc()
```

### Global Scope

A variable created in the main body of the Python code is a global variable and belongs to the global scope.

Global variables are available from within any scope, global and local.

```python
x = 300

def myfunc():
    print(x)

myfunc()

print(x)
```

### Naming Variables

If you operate with the same variable name inside and outside of a function, Python will treat them as two separate variables, one available in the global scope (outside the function) and one available in the local scope (inside the function)

Below examples prints the local x first, and then the code will print the global x:

```python
x = 300
def myfunc():
    x = 200
    print(x)

myfunc()
print(x)
```

### Global Keyword

If you need to create a **global** variable, but are stuck in the **local** scope, you can use the **global** keyword.

The **global** keyword makes the variable global.



```python
def myfunc():
    global x
    x = 300

myfunc()
print(x)
```

```python
x = 300
print(x)

def myfunc():
    global x
    x = 200

myfunc()

print(x)
```

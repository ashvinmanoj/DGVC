### pandas.DataFrame
**A pandas DataFrame can be created using the following constructor**

*pandas.DataFrame( data, index, columns, dtype, copy)*

**data**
*data takes various forms like ndarray, series, map, lists, dict, constants and also another DataFrame.*

**index**
*For the row labels, the Index to be used for the resulting frame is Optional Default np.arange(n) if no index is passed.*


**Columns**
*For column labels, the optional default syntax is - np.arange(n). This is only true if no index is passed.*

**dtype**
*Data type of each column.*

**copy**
*This command (or whatever it is) is used for copying of data, if the default is False.*

#### Selecting Data  from datafarme

When it comes to select data on a DataFrame, Pandas loc and iloc are two top favorites. They are quick, fast, easy to read, and sometimes interchangeable.


* Differences between loc and iloc
* Selecting via a single value
* Selecting via a list of values
* Selecting a range of data via slice
* Selecting via conditions and callable
* loc and iloc are interchangeable when labels are 0-based integers

### 1. Differences between loc and iloc
The main distinction between loc and iloc is:

loc is label-based, which means that you have to specify rows and columns based on their row and column labels.
iloc is integer position-based, so you have to specify rows and columns by their integer position values (0-based integer position).

### 2. Selecting via a single value

```python
data.loc['JF320208', 'Primary Type']
```

#### The equivalent `iloc` for above statement should take row number 4 and column number 1# The equivalent `iloc` statement should take row number 4 and column number 1

```python
data.iloc[2, 4]
```

#### To get all rows

```python
data.loc[:, 'Description']
```

#### The equivalent `iloc` statement

```python
data.iloc[:, 5]
```

#### To get all columns

```python
data.loc['JF310109', :]
```

```python
type(data.loc['JF310109', :])
```

### 3. Selecting via a list of values

```python
data.tail()
```

#### Multiple rows using iloc

#### What is the datatype of the multi row and multi column selection for loc and iloc

```python
rows = ['JF294494', 'JF317539']
cols=['Description','Location Description']
print(type(data.loc[rows, cols]))

rows = [215548, 215550]
cols = [5, 6]
print(type(data.iloc[rows, cols]))
```

### 4. Selecting a range of data via slice

```python
# Slicing column labels
rows= ['JF294494', 'JF317539']
data.loc[rows, 'Description':'Year' ]
```

```python
# Slicing row labels
cols = ['Description', 'Year']
data.loc['JF315350':'JF313061', cols]
```

#### We can use the syntax A:B:S to select data from label A to label B with step size S (Both A and B are included):

```python
data.loc['JF315350':'JF313061':2, cols]
```

#### With iloc, we can also use the syntax n:m to select data from position n (included) to position m (excluded).

```python
data.iloc[[1, 2], 0 : 3]
```

### 5. Selecting via conditions and callable
#### 5.2 Conditions

```python
# One condition
data.loc[data.Beat < 999  , :]
```

```python
## multiple conditions
data.loc[(data.Beat <999) & (data.IUCR < '200'), ['Description', 'Domestic']         ]
```

#### Getting ValueError

```python
data.iloc[data.Beat <999, :]
```

#### Fix value error

```python
data.iloc[list(data.Beat <999), :]
```

### Multiple conditions

```python
## multiple conditions
data.iloc[ list((data.Beat <999) & (data.IUCR < '200')),     :  ]
 
```

### 5.2 Callable

```python
# Selecting columns
data.loc[:, lambda df:  ['Description', 'Domestic'] ]
```

### With condition

```python
data.loc[lambda data: data.Beat <999, :]
```

```python
data.iloc[lambda data: [0,1], :]
```

```python
data.iloc[lambda data: list(data.Beat < 200), :]
```

### 6. loc and iloc are interchangeable when labels are 0-based integers

```python
 
url2= "https://raw.githubusercontent.com/RandomFractals/chicago-crimes/main/data/crimes-2022.csv"
data=pd.read_csv(url2,header=None )
data['Row Number']= data.index
data
```

```python
data.loc[1, 2]
```

```python
data.loc[1, [1, 2,3]]
```

```python
data.loc[1, 2] == data.iloc[1, 2]
```

```python
data.loc[1, [1, 2]] == data.iloc[1, [1, 2]]
```

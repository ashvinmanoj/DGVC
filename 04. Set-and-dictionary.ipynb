### Built-in Data Types

 

### Python Collections (Arrays)

There are four collection data types in the Python programming language:

**List** is a collection which is ordered and changeable. Allows duplicate members.

**Tuple** is a collection which is ordered and unchangeable. Allows duplicate members.

**Set** is a collection which is unordered, unchangeable*, and unindexed. No duplicate members.

**Dictionary** is a collection which is ordered** and changeable. No duplicate members.

*Set items are unchangeable, but you can remove items and add new items.

**As of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered.

### SET

**Adding elements to a set**

```python
Set_variable = {"apple", "banana", "cherry"}
print("Set_variable before addition of a new element :", Set_variable)
Set_variable.add("orange")
print("Set_variable after addition of a new element :", Set_variable)
```

**Updating a set object**

```python
Set_variable1 = {"apple", "banana", "cherry"}
Set_variable2 = {"pineapple", "mango", "papaya"}

Set_variable1.update(Set_variable2)

print(Set_variable1)
```

**Updating a set with a list object**

```python
Set_variable1 = {"apple", "banana", "cherry"}
List_variable = ["kiwi", "orange"]

Set_variable1.update(List_variable)

print(Set_variable1)
```

**Deleting an element from a SET**
 
**To remove/delete an item in a set, use the remove(), or the discard() method or pop() method.**

Sets are unordered, so when using the pop() method, you do not know which item that gets removed.

```python
Set_variable = {"apple", "banana", "cherry"}

x = Set_variable.pop()

print(x)

print(Set_variable)
```

Using **Discard**- If the item to remove does not exist, discard() will NOT raise an error.

```python
Set_variable = {"apple", "banana", "cherry"}

Set_variable.discard("banana")
print(Set_variable)
Set_variable.discard("mango")
print(Set_variable)
```

If the item to remove in a set object does not exist, **remove() will raise an error.**

```python
Set_variable = {"apple", "banana", "cherry"}

Set_variable.remove("mango")

print(Set_variable)
```

## Dictionary

Dictionaries are used to store data values in key:value pairs.

A dictionary is a collection which is ordered*, changeable and do not allow duplicates.

*   **As of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered.*

```python
Dict_variable = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(dict_variable)
```

```python
Dict_variable = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(Dict_variable)
```

**Dictionary items are presented in key:value pairs, and can be referred to by using the key name.**

```python
Dict_variable = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(Dict_variable["brand"])
```

**Duplicates Not Allowed**

```python
Dict_variable = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "year": 2020
}
print(Dict_variable)
```

**Update Dictionary using the update() method**

```python
Dict_variable = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
Dict_variable.update({"year": 2020})
print(Dict_variable)
```

```python
Dict_variable = dict(name = "John", age = 36, country = "Norway")
print(Dict_variable)
```

**Access Dictionary Items**

```python
Dict_variable = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = Dict_variable["model"]
```

**Get keys in a dictionary**

```python
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}

dict_keys = car.keys()

print(dict_keys) #before the change

car["color"] = "white"

print(dict_keys) #after the change
```

**Get items (Key:value pairs) in a dictionary**

```python
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}

dict_keys = car.items()

print(dict_keys) #before the change

car["year"] = 2020

print(dict_keys) #after the change
```

**Access an item from a dictionary**

```python
dict_variable = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = dict_variable["model"]
print(x)
```

**Access an item from a dictionary using get method**

```python
x = dict_variable.get("model")
print(x)
```

**How to get Keys and Values seperately from a dictionary**

```python
 car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}


print(car.keys())
print(car.values())
```

**The pop() method removes the item with the specified key name:**

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.pop("model")
print(thisdict)
```

**The popitem() method removes the last inserted item**

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.popitem()
print(thisdict)
```

**The del keyword removes the item with the specified key name**

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
del thisdict["model"]
print(thisdict)
```
